Importamos librerias

```python
import pandas as pd
from sqlalchemy import create_engine
```

Leemos archivo csv

```python
df = pd.read_csv('AccidentesAviones.csv')
```

```python
pd.options.display.max_columns = None # Con esta linea de código vemos todas las columnas del DataFrame
```

```python
df.head(2)
```

Dropeamos las columnas que no vamos a usar en el dashboard

```python
df2 = df.drop(columns = ['Unnamed: 0', 'HORA declarada', 'flight_no', 'route', 'registration', 'cn_ln', 'PASAJEROS A BORDO', 'crew_aboard', 'passenger_fatalities', 'crew_fatalities', 'summary'])
df2
```

```python
df2['fecha'] = pd.to_datetime(df2.fecha, errors = 'coerce') # Cambiamos la codificación de la columna 'fecha'
```

```python
df2
```

Reemplazamos los signos '?' por un '0' transformando los valores de las columnas a string

```python
df2['fecha'] = df2['fecha'].astype('string').str.replace('?','0')
df2['Ruta'] = df2['Ruta'].astype('string').str.replace('?','0')
df2['OperadOR'] = df2['OperadOR'].astype('string').str.replace('?','0')
df2['ac_type'] = df2['ac_type'].astype('string').str.replace('?','0')
df2['all_aboard'] = df2['all_aboard'].astype('string').str.replace('?','0')
df2['cantidad de fallecidos'] = df2['cantidad de fallecidos'].astype('string').str.replace('?','0')
df2['ground'] = df2['ground'].astype('string').str.replace('?','0')
```

```python
df2
```

Creamos un nuevo archivo .csv con las columnas seleccionadas y valores modificados

```python
df2.to_csv('AccidentesAviones2.csv', sep=',',index=False)
```
